```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation, rc
from IPython.display import HTML, Image
rc('animation',html = 'html5')
data = np.genfromtxt("data_EOB_RK4.txt")
r = data[:50000,0]
theta = data[:50000,1]
phi = data[:50000,2]
%pylab inline
x = r*np.sin(theta)*np.cos(phi)
y = r*np.sin(theta)*np.sin(phi)
z = r*np.cos(theta)

empty = np.zeros(len(x))
fig = plt.figure()
ax = Axes3D(fig)
ax.plot(x[:],y[:],z[:],label="circular orbit")
ax.plot(x[:],y[:],empty-9,alpha=0.2)
ax.plot(x[:],empty+9,z[:],alpha=0.2)
ax.plot(empty-9,y[:],z[:],alpha=0.2)
distance = np.sqrt(x**2+y**2+z**2)
plt.savefig("orbit",dpi=500)
plt.show()
```
